```python
#Purchasing Analysis (Gender)
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

# PURCHASING ANALYSIS (GENDER)
# Calculate purchase count by gender
Gender_Group_df = purchase_data.groupby('Gender')

Number_Players = Gender_Group_df['SN'].count()
Number_Players_df = pd.DataFrame(Number_Players)

# Calculate average purchase price by gender
Average_Price_by_Gender = Gender_Group_df['Price'].mean()

# Calculate total purchase price by gender
Total_Gender_Purchases = Gender_Group_df['Price'].sum()

# Calculate normalized totals (total per category / total count per category)
Normal_Total = Total_Gender_Purchases / Number_Players

# Add 'Average Purchase Price' to table
Number_Players_df['Average Purchase Price'] = Average_Price_by_Gender
Gender_Purchases_df = pd.DataFrame(Number_Players_df)

# Rename 'SN' column to 'Purchase Count'
Update_Gender_Purchases_df = Gender_Purchases_df.rename(columns={
    'SN': 'Purchase Count',
})

# Add 'Normalized Totals' to table
Update_Gender_Purchases_df['Total Purchase Price'] = Total_Gender_Purchases
Updated1_Gender_Purchases_df = pd.DataFrame(Update_Gender_Purchases_df)

# Add 'Total Purchase Price' to table
Updated1_Gender_Purchases_df['Normalized Totals'] = Normal_Total
Total_Gender_Purchases_df = pd.DataFrame(Updated1_Gender_Purchases_df)
Total_Gender_Purchases_df

```
